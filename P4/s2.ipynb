```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fi = [100, 1e3, 10e3, 25e3, 40e3, 75e3, 100e3, 500e3, 1e6]
vo = [512e-3, 7.44, 8.64, 6.68, 3.64, 2.42, 110e-3, 22e-3, 8e-3]
vi = [20e-3, 20e-3, 20e-3, 20e-3, 20e-3, 20e-3, 20e-3, 20e-3, 20e-3]
g  = [a/b for a, b in zip(vo, vi)]
gdb= [20 * np.log(i) for i in g]
```

```python
for i in gdb:
    print("\\qty{" + f"{int(i/2 * 100) / 100}"+ "}{\\deci\\bel}", end=" & ")
```

```python
fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(fi, gdb)
ax.set_xlabel("Freqncia d\'entrada [Hz]")
ax.set_ylabel("Guany [dB]")
ax.set_xscale("log")
ax.grid()
ax.minorticks_on()
ax.grid(True, which="minor", axis="both", linestyle="--")
plt.savefig("s2_graph.pdf", )
```
