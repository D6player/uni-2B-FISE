```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fi = [100, 200, 500, 750, 900, 1e3, 1.2e3, 1.5e3, 2e3, 5e3, 10e3]
vipp = [400e-3 for _ in fi]
vopp = [88e-3, 168e-3, 512e-3, 1.26, 2.84, 3.7, 1.7, 840e-3, 484e-3, 168e-3, 86e-3]
g  = [a/b for a, b in zip(vopp, vipp)]
gdb= [20 * np.log(i) for i in g]
```

```python
for i in gdb:
    print("\\qty{" + str(int(i * 100)/100) + "}{\\deci\\bel}", end=" & ")
```

```python
fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(fi, gdb)
ax.set_xlabel("Freqncia d\'entrada [Hz]")
ax.set_ylabel("Guany [dB]")
ax.set_xscale("log")
ax.grid()
ax.minorticks_on()
ax.grid(True, which="minor", axis="both", linestyle="--")
plt.savefig("fotos/s2_graph.pdf", )
```
